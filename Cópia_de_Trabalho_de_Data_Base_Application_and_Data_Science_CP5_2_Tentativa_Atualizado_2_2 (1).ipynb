## 2015 World Happiness Report

O World Happiness Report é uma pesquisa histórica sobre o estado da felicidade global

- País - Nome do país
- Região - Região à qual o país pertence
- Classificação de felicidade - Classificação do país com base no Índice de Felicidade.
- Pontuação de Felicidade - Uma métrica medida em 2015 perguntando às pessoas amostradas: "Como você classificaria sua felicidade em uma
- Família - A medida em que a Família contribui para o cálculo do Índice de Felicidade
- Saúde (Expectativa de Vida) - A extensão em que a esperança de vida contribuiu para o cálculo do Índice de Felicidade
- Liberdade Até que ponto a Liberdade contribuiu para o cálculo do Índice de Felicidade.
- Confiança (Corrupção Governamental) - A extensão em que a Percepção da Corrupção contribui para a Pontuação de Felicidade.

As pontuações e classificações de felicidade usam dados da Gallup World Poll. As pontuações são baseadas em respostas à principal pergunta de avaliação de vida feita na pesquisa. Esta pergunta, conhecida como a escada de Cantril, pede aos entrevistados que pensem em uma escada com a melhor vida possível para eles sendo 10 e a pior vida possível sendo 0 e classifiquem suas próprias vidas atuais nessa escala.

### Exploração

```python
# Calcular a média, mediana e moda da Pontuação de Felicidade
media_pontuacao = df_2015['Pontuação de Felicidade'].mean()
mediana_pontuacao = df_2015['Pontuação de Felicidade'].median()
moda_pontuacao = df_2015['Pontuação de Felicidade'].mode()[0]  # Pega o primeiro valor da moda

# 2. Visualização da Distribuição da Pontuação de Felicidade
plt.figure(figsize=(10, 6))
sns.histplot(df_2015['Pontuação de Felicidade'], bins=15, kde=True)
plt.title('Distribuição da Pontuação de Felicidade em 2015')
plt.xlabel('Pontuação de Felicidade')
plt.ylabel('Frequência')

# Adicionar média, mediana e moda ao gráfico
plt.axvline(media_pontuacao, color='red', linestyle='--', label=f'Média: {media_pontuacao:.2f}')
plt.axvline(mediana_pontuacao, color='blue', linestyle='--', label=f'Mediana: {mediana_pontuacao:.2f}')
plt.axvline(moda_pontuacao, color='green', linestyle='--', label=f'Moda: {moda_pontuacao:.2f}')
plt.legend()

# Exibir o gráfico
plt.show()

```

```python
#Análise de Correlação
plt.figure(figsize=(12, 8))
correlation_matrix = df_2015[numeric_columns].corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Matriz de Correlação')
plt.show()
```

Muitas áreas rurais têm acesso limitado a cuidados de saúde, água potável e saneamento adequado, o que pode contribuir para uma menor expectativa de vida.

4. Nutrição e insegurança alimentar
Embora menos grave do que em outras partes da África, a insegurança alimentar em Botsuana ainda é uma questão em algumas áreas. A desnutrição infantil e a insegurança alimentar crônica podem prejudicar o desenvolvimento e aumentar a mortalidade infantil, afetando a expectativa de vida global.


O Paradoxo de Easterlin, proposto pelo economista Richard Easterlin em 1974, sugere que, embora a renda de um país aumente ao longo do tempo, isso não resulta em um aumento proporcional da felicidade média da população.

Atualizações e Discussões Recentes:
Estudos mais recentes têm aprofundado o paradoxo, mostrando que o aumento de renda tem um efeito limitado na felicidade.

Em resumo, a ideia central é que, após um certo nível de bem-estar econômico, mais dinheiro não resulta necessariamente em mais felicidade, e outros aspectos da vida passam a ser mais relevantes para o bem-estar.

3. Mudança de Expectativas
Quando um país cresce economicamente, as expectativas da população também aumentam. Se o crescimento não acompanhar a melhoria na qualidade de vida geral (por exemplo, serviços de saúde, educação, segurança), isso pode gerar frustração, mesmo com indicadores econômicos e de saúde melhores.

4. Impactos de Eventos Negativos
Além disso, eventos como crises políticas, choques econômicos ou epidemias podem ter impactos negativos sobre o bem-estar psicológico das pessoas, independentemente de melhorias em outros setores, como saúde ou economia.
